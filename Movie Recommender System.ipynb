```python
import ast
def convert(list_of_dict):
    return [i['name'] for i in ast.literal_eval(list_of_dict)]
```

```python
movie_df['genres'] = movie_df['genres'].apply(convert)
movie_df['keywords'] = movie_df['keywords'].apply(convert)
```

```python
import ast
def convert_cast(list_of_dict):
    return [i['name'] for i in ast.literal_eval(list_of_dict)][0:3]

movie_df['cast'] = movie_df['cast'].apply(convert_cast)
```

```python
import ast
def convert_cast(list_of_dict):
    return [i['name'] for i in ast.literal_eval(list_of_dict) if i['job'] == 'Director' ]

movie_df['crew'] = movie_df['crew'].apply(convert_cast)
```

```python
movie_df['overview'] = movie_df['overview'].apply(lambda x:x.split())
```

```python
movie_df['genres'] = movie_df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movie_df['keywords'] = movie_df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movie_df['overview']=movie_df['overview'].apply(lambda x:[i.replace(" ","") for i in x])
movie_df['cast'] = movie_df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movie_df['crew'] = movie_df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie_df['tags'] = movie_df['overview']+movie_df['genres']+movie_df['keywords']+movie_df['cast']+movie_df['crew']
```

```python
mov_df = movie_df[['movie_id','title','tags']]
mov_df.head()
```

```python
mov_df['tags'] = mov_df['tags'].apply(lambda x:" ".join(x))
mov_df['tags'][0]
```

```python
mov_df['tags'] = mov_df['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(mov_df['tags']).toarray()
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    stemmed_words = [ps.stem(i) for i in text.split()]
    return " ".join(stemmed_words)
        
```

```python
mov_df['tags'] = mov_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    mov_index = mov_df[mov_df['title']==movie].index[0]
    distances = similarity[mov_index]
    movies_list = sorted(list(enumerate(distances)),key=lambda x:x[1],reverse=True)[1:6]
    return [mov_df.iloc[i[0]].title for i in movies_list]
```

```python
recommend('Batman Begins')
```
